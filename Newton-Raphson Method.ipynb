# Newton-Raphson Method

## Introduction
This is a Python code that implements Newton-Raphson Method to find the approximation value of the root of a nonlinear equation. If you do not know what Newton-Raphson Method is or how it works, you can check this following [link](https://brilliant.org/wiki/newton-raphson-method/). ## Example
Find the root of $ f(x) = 5x^3 - 2x^2 - 3x $ with $ x_0 = 6 $ and $ ε = 10^-6 $.

## Import Libraries
Import the libraries that are needed.

```python
import numpy as np
import pandas as pd
```

## Equation
Defining our equation.

```python
def f(x):
    return 5*x**3 - 2*x**2 - 3*x
```

## Derivative of the Equation
Defining the derivative of our equation.

```python
def f_aksen(x):
    return 15*x**2 - 4*x - 3
```

## Newton-Raphson
Implementing the Newton-Raphson Method to solve our equation.

```python
def newton_raphson(x0,error,N):
    print("\nIteration Table")
    n = 0
    # Create empty lists that we're going to insert our answers into
    list1 = []
    list2 = []
    list3 = []
    list4 = []
    list5 = []
    # Set conditions that should be met
    convergent = True
    condition = True

    while condition:
        # If the derivative of our equation is equal to zero, stop the calculation process
        if f_aksen(x0) == 0:
            print("\nDivision by a number close to 0")
            break
        
        # Insert our answers to the empty lists that we made earlier
        while n == 0:
            xn = x0
            list1.append(n)
            list2.append(xn)
            list3.append(f(xn))
            list4.append(f_aksen(xn))
            list5.append(np.nan)
            n += 1

        xn = x0 - (f(x0)/f_aksen(x0))
        difference = abs(xn - x0)
        list1.append(n)
        list2.append(xn)
        list3.append(f(xn))
        list4.append(f_aksen(xn))
        list5.append(difference)
        x0 = xn
        n += 1
        # If the numbers of iterations exceed the maximum iterations, we can assume that our equation is divergent
        if n > N:
            convergent = False
            break
        condition = difference > error
    
    # Insert the lists that contain the answers into a dataframe, so that the output will look better
    df = pd.DataFrame({"n":np.round(np.array(list1),6),"xn":np.round(np.array(list2),6),"f(xn)":np.round(np.array(list3),6),
        "f'(xn)":np.round(np.array(list4),6),"|xn+1 - xn|":np.round(np.array(list5),6)}).set_index("n")
    
    # If all of the conditions are met, print our dataframe 
    if convergent == True:
        print("\n",df)
        print("\nDue to |x{} - x{}| < {}, it can be assumed that the approximation value of the root is {}".
            format(n - 1,n - 2,error,round(xn,6)))
    # If not, print this
    else:
        print("\nThis equation is divergent")

# Input the values of x0, tolerable error, and maximum iterations
x0 = float(input("x0: "))
error = float(input("Tolerable Error: "))
N = float(input("Maximum Iterations: "))

newton_raphson(x0,error,N)
```
